```python
# 1. 괄호 & 공백 없애기
gs_open_name_list = gs_open['사업장명']
def remove_braket():
  remove_list = []
  for i in gs_open_name_list:
    remove_list.append(i.replace( "(" , "").replace(')', "").replace(" ",""))
  return remove_list
gs_name_list_1 = remove_braket()
gs_name_list_1[:5]
# 제거 - 레즐리컵LESELYCUP,컵CUP커피,서울식물원푸드이마트24편의점스무디킹, 카페커디CAFECUDI
# 변환 - 지에스25GS25S김포공항역점,지에스25시영일점,주이마트24강서BSD센터CFF점,주이마트24김포공항
```

```python
remove_shop_list = ['레즐리컵LESELYCUP','컵CUP커피','서울식물원푸드이마트24편의점스무디킹', '카페커디CAFECUDI']
remove_idx = [e for i in remove_shop_list for e,x in enumerate(gs_name_list_1) if i in x]
display(gs_open.iloc[remove_idx])
gs_open.drop(remove_idx, inplace=True) # 삭제
gs_open = gs_open.reset_index()
```

```python
# 도로명 주소 -> 위도,경도로 변환을 위해 주소 수정 
import re
def modify_address():
  pattern = r'\([^)]*\)' # 괄호 안 내용 삭제
  temp = [re.sub(pattern=pattern, repl='', string= x) for x in gs_open['도로명주소']]
  temp_split = [x.split(',')[0] for x in temp]
  temp_remove = [x.replace(')',"") for x in temp_split]# 괄호 남아있는 거 삭제
  new_address = [x.rstrip() for x in temp_remove] # 왼쪽 공백 삭제
  return new_address
gs_open['도로명주소'] = modify_address()
gs_open['도로명주소'].head()
```

```python
from tqdm import tqdm, tqdm_pandas

tqdm.pandas()
```

```python
from geopy.geocoders import Nominatim
geo_local = Nominatim(user_agent='South Korea')

#위도 반환 함수
def geocoding_lat(address):
    try:
        geo = geo_local.geocode(address)
        return geo.latitude

    except:
        return 0

#경도 반환 함수
def geocoding_lon(address):
    try:
        geo = geo_local.geocode(address)
        return geo.longitude

    except:
        return 0
        
# 변환
gs_open['위도'] = gs_open['도로명주소'].progress_apply(geocoding_lat)
gs_open['경도'] = gs_open['도로명주소'].progress_apply(geocoding_lon)
```

```python
from geopy.geocoders import Nominatim
geo_local = Nominatim(user_agent='South Korea')

#위도 반환 함수
def geocoding_lat(address):
    try:
        geo = geo_local.geocode(address)
        return geo.latitude

    except:
        return 0

#경도 반환 함수
def geocoding_lon(address):
    try:
        geo = geo_local.geocode(address)
        return geo.longitude

    except:
        return 0
        
# 변환
gs_con['위도'] = gs_con['도로명주소'].progress_apply(geocoding_lat) # 기존의 편의점 데이터의 위도 경도와 geopy를 통해 변환한 위도 경도의 오차가 있어 도로명 주소를 기준으로 동일하게 변경해줌
gs_con['경도'] = gs_con['도로명주소'].progress_apply(geocoding_lon)
```
