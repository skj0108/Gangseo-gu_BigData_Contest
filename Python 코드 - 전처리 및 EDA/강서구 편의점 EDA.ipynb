### 분석에서 활용할 수 있는 정보
##### 상호명, 지점명, 행정동명, 법정동명, 도로명주소, 층정보, 경도, 위도
- 상호명, 지점명: 웹 검색을 통한 실제 편의점의 정보 확인 가능
- 행정동명, 법정동명: 동별 편의점 통계 확인 가능
- 층정보: 장애인들이 사용하기에 편리한지의 여부
- 도로명주소, 경도, 위도: 지도에 시각화 시, 활용 가능

```python
store_need = store[['상호명','지점명','행정동명','법정동명','도로명주소','층정보','경도','위도']]
```

```python
store_need
```

```python
# 상호명이 결측치인 행은 없으며,
# 지점명이 결측치인 행은 58개이며, 상호명이 지점명으로 되어있거나 프랜차이즈의 이름만 나와있는 경우이다.

display(store_need[store_need['상호명'].isnull()])
display(store_need[store_need['지점명'].isnull()])
```

```python
# 주소 관련 컬럼들은 결측치를 가지고 있는 행이 없다.

display(store_need[store_need['행정동명'].isnull()])
display(store_need[store_need['법정동명'].isnull()])
display(store_need[store_need['도로명주소'].isnull()])
```

```python
# 층정보 컬럼이 결측치인 행은 95개이다.

display(store_need[store_need['층정보'].isnull()])
```

```python
# 경도, 위도 컬럼이 결측치인 행은 없다.

display(store_need[store_need['경도'].isnull()])
display(store_need[store_need['위도'].isnull()])
```

```python
# 법정동별로 편의점의 개수를 보았을 때, 화곡동/마곡동/등촌동 순으로 개수가 많은 것을 알 수 있다

store_need.groupby(['법정동명']).count().상호명.sort_values(ascending = False)
```

```python
# 행정동을 기준으로 보면, 가양1동/화곡1동/발산1동 순으로 개수가 많은 것을 알 수 있다

store_need.groupby(['행정동명']).count().상호명.sort_values(ascending = False)
```

```python
store_need.지점명.fillna('',inplace=True)
```

```python
location_data = store_need[['위도','경도']].values[:len(store_need)].tolist()
center =[store_need.위도.mean(),store_need.경도.mean()]
```

```python
map = folium.Map(location=center,zoom_start=13,width='70%')
```

```python
for i in range(len(store_need)):
    store_name=(str(store_need['상호명'][i]+' '+store_need['지점명'][i]))
    store_color='black'
    folium.Circle(location=location_data[i],radius=1,color=store_color,fill=True,fill_opacity=1, tooltip=store_name).add_to(map)

plugins.HeatMap(location_data,radius=18).add_to(map)
```
