```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import folium
from folium import plugins

dis_protect = pd.read_csv('dis_protect.csv', encoding = 'CP949')
dis_job = pd.read_csv('dis_job.csv', encoding = 'CP949')
dis_rehab = pd.read_csv('dis_rehab.csv', encoding = 'CP949')
dis_resident = pd.read_csv('dis_resident.csv', encoding = 'CP949')
```

```python
dis_protect['시설종류'] = '노인장애인보호구역'
dis_protect = dis_protect[['대상시설명','시설종류','위도','경도']]
dis_protect.columns = ['시설명','시설종류','위도','경도']
```

```python
dis_job = dis_job[['시설명','시설종류상세명(시설종류)','위도','경도']]
dis_job.columns = ['시설명','시설종류','위도','경도']
```

```python
dis_rehab = dis_rehab[['시설명','시설종류상세명(시설종류)','위도','경도']]
dis_rehab.columns = ['시설명','시설종류','위도','경도']
dis_rehab.loc[3,'위도'] = 37.574191
dis_rehab.loc[3,'경도'] = 126.812563
```

```python
dis_resident = dis_resident[['시설명','시설종류상세명(시설종류)','위도','경도']]
dis_resident.columns = ['시설명','시설종류','위도','경도']
```

```python
facility = pd.concat([dis_protect, dis_job, dis_rehab, dis_resident], ignore_index=True)
```

```python
facility
```

```python
location_data = facility[['위도','경도']].values[:len(facility)].tolist()
center =[facility.위도.mean(),facility.경도.mean()]
```

```python
map = folium.Map(location=center,zoom_start=13,width='100%')
for i in range(len(facility)):
    facility_name=facility.시설명[i]
    facility_color='black'
    folium.Circle(location=location_data[i],radius=1,color=facility_color,fill=True,fill_opacity=1, tooltip=facility_name).add_to(map)

plugins.HeatMap(location_data,radius=20).add_to(map)
```
