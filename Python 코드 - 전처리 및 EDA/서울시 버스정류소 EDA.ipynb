### 분석에서 활용할 수 있는 정보
##### 정류소번호, 정류소명, X좌표, Y좌표
- 정류소번호가 16으로 시작하며 5자리인것은 강서구에 위치한 정류소.
- X좌표, Y좌표를 통해 강서구 내의 버스 정류소를 시각화.
- 시각화를 통해 버스정류소의 분포 확인 가능.

```python
# 정류소번호로 강서구에 위치한 정류소를 골라내기 위해 문자열로 변환.
bus = bus.astype({'정류소번호':'str'})
```

```python
# 정규표현식을 통해 정류소번호가 16으로 시작하고 5자리인 것을 필터링하여 변수에 저장.
bus_gs = bus[bus.정류소번호.str.contains('16+[0-9]+[0-9]+[0-9]+')]
bus_gs.reset_index(inplace=True)
```

```python
location_data = bus_gs[['Y좌표','X좌표']].values[:len(bus_gs)].tolist()
center =[bus_gs.Y좌표.mean(),bus_gs.X좌표.mean()]
map = folium.Map(location=center,zoom_start=13,width='70%')
```

```python
for i in range(len(bus_gs)):
    bus_name=bus_gs['정류소명'][i]
    bus_color='black'
    folium.Circle(location=location_data[i],radius=1,color=bus_color,fill=True,fill_opacity=1, tooltip=bus_name).add_to(map)

plugins.HeatMap(location_data,radius=20).add_to(map)
```
