### 데이터 전처리

```python
con_1.상호명.unique() # CU, 세븐일레븐, GS25, 미니스톱, 이마트24을 대상으로 추출
```

```python
#상호명에 CU,세븐일레븐,GS25,이마트24,지에스25,미니스톱 가 포함되는 데이터만 추출
gs_con = con_1[con_1.상호명.str.contains('CU|세븐일레븐|GS25|지에스25|이마트24|미니스톱')]
gs_con.head()
```

```python
gs_con.상호명.unique()
```

```python
# 상호명 잘못 저장된 데이터 list에 저장
con_name = gs_con.상호명.unique().tolist()
con_name = [name for name in con_name if name not in ['GS25','CU','이마트24','세븐일레븐','미니스톱']]
```

```python
gs_con[gs_con.상호명.isin(con_name)].head(5)
```

1. 상호명에 지점명이 포함된 경우 - 상호명만 남기기

```python
replace_df = gs_con[gs_con.상호명.isin(con_name)] # 수정이 필요한 데이터
def replace_name():
  replace_name_list = []
  for i in replace_df.상호명:
    if 'CU' in i:
      replace_name_list.append('CU')
    elif 'GS25' in i:
      replace_name_list.append('GS25')
    elif '지에스25' in i:
      replace_name_list.append('GS25')
    elif '세븐일레븐' in i:
      replace_name_list.append('세븐일레븐')
    elif '이마트24' in i :
      replace_name_list.append('이마트24')
    elif '미니스톱' in i :
      replace_name_list.append('미니스톱')
    else:
      replace_name_list.append(i)
  return replace_name_list
replace_name_list=  replace_name()
```

```python
np.unique(replace_name_list)
```

```python
gs_con.loc[replace_df.index,'상호명'] = replace_name_list
gs_con.head(6)
```

```python
gs_con.상호명.unique()
```

```python
temp_name = ['방화또오기점','화곡풀향기점','가양프라자점','화곡달빛점','발산1단지점','뉴마곡14단지점','염창세빛점',
 '등촌캐빈점','마곡골든점','방화샤르망점','화곡대원점']
# 해당 위치에 편의점 없는 경우 : 26,100,146,184,196,226,231
not_con_idx = [26,100,146,184,196,226,231]
```

```python
gs_con[gs_con.지점명.isnull()].index
```

```python
temp_idx = gs_con[gs_con.지점명.isnull()].index
temp_idx= temp_idx.drop(not_con_idx)
temp_idx
```

```python
gs_con = gs_con.drop(not_con_idx)
gs_con.loc[temp_idx,'지점명'] = temp_name
```

```python
# 중복 데이터 확인 및 제거
gs_con[gs_con.duplicated(['상호명','지점명'], keep = False)] # 확인
```

```python
# 중복 1개씩 제거, 세븐일레븐 마곡마커스점 실제 존재 X 2개 모두 제거
drop_idx = [24,29,89,199,286,317,342]
gs_con = gs_con.drop(drop_idx)
gs_con.reset_index(drop=True,inplace = True)
```
