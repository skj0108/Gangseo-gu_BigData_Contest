```python
import pandas as pd
import matplotlib.pyplot as plt
import pymysql
import folium
from folium import plugins
```

```python
charge = pd.read_csv('서울시 전동보장구 급속충전소.csv')
charge.head()
```

```python
charge_gs = charge[charge['시군구명'] == '강서구']
charge_gs.reset_index(inplace=True)
```

```python
charge_gs = charge_gs[['시설명', '시군구명', '소재지도로명주소', '위도', '경도', '설치장소설명', '동시사용가능대수', '공기주입가능여부', '휴대전화충전가능여부']]
```

```python
charge_gs.head()
```

```python
charge_gs.to_csv('강서구_전동보장구_위치.csv', index=False, encoding="utf-8-sig")
```

```python
location_data = charge_gs[['위도','경도']].values[:len(charge_gs)].tolist()
center =[charge_gs.위도.mean(),charge_gs.경도.mean()]
map = folium.Map(location=center,zoom_start=13,width='70%')
```

```python
for i in range(len(charge_gs)):
    charge_name=charge_gs.시설명[i]
    charge_color='black'
    folium.Circle(location=location_data[i],radius=1,color=charge_color,fill=True,fill_opacity=1, tooltip=charge_name).add_to(map)

plugins.HeatMap(location_data,radius=20).add_to(map)
```
